```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
```

```python
file_path = "../data/raw/mobile_app_usage.csv"

df_raw = pd.read_csv(file_path)

print("Raw Dataset Shape:", df_raw.shape)
display(df_raw.head())
```

```python
selected_columns = [
    "record_id",
    "year",
    "month",
    "day_of_week",
    "country",
    "age_group",
    "gender",
    "app_category",
    "app_name",
    "subscription_type",
    "daily_screen_time_minutes",
    "session_duration_minutes",
    "sessions_per_day",
    "app_opens_per_day",
    "notifications_received_per_day",
    "notification_settings",
    "primary_usage_time",
    "sleep_disruption_from_phone",
    "screen_time_concern",
    "mental_health_impact",
    "digital_wellbeing_feature_used"
]

df = df_raw[selected_columns].copy()

print("Analysis Dataset Shape:", df.shape)
display(df.head())
```

```python
print("Duplicate Rows:", df.duplicated().sum())

quality_audit = pd.DataFrame({
    "dtype": df.dtypes.astype(str),
    "missing_count": df.isnull().sum(),
    "missing_pct": (df.isnull().mean() * 100).round(2),
    "unique_values": df.nunique()
})

display(quality_audit)
```

```python
df["sleep_disruption_from_phone"] = (
    df["sleep_disruption_from_phone"]
    .fillna("Not Reported")
)

print(df["sleep_disruption_from_phone"].value_counts())
```

```python
numeric_cols = [
    "daily_screen_time_minutes",
    "session_duration_minutes",
    "sessions_per_day",
    "app_opens_per_day",
    "notifications_received_per_day"
]

display(df[numeric_cols].describe().T.round(2))
```

```python
output_path = "../data/processed/screensense_clean.csv"

df.to_csv(output_path, index=False)

print("Cleaned dataset saved.")
print("Final Shape:", df.shape)
print("Total Missing Values:", df.isnull().sum().sum())
```
